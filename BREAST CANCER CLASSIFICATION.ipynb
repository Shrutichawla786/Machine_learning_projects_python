Splitting the data into tranning and testing data

```python
x_train , x_test , y_train , y_test= train_test_split(x , y , test_size=0.2 , random_state=2)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

Model Training 

Logistic Regression


```python
algo =LogisticRegression()
```

```python
algo.fit(x_train , y_train)
```

```python
y_train_predict=algo.predict(x_train)
y_test_predict= algo.predict(x_test)
```

```python
plt.plot(y_train , y_train_predict)
```

```python
plt.plot(y_test , y_test_predict)
```

```python
train_score=algo.score(x_train , y_train)
print("Acuracy on traning data : " , train_score)
```

```python
test_score=algo.score(x_test , y_test)
print("Acuracy on testing data : " ,test_score )
```
